### Best Years

In 1981, Dennis Martinez was leading the league in wins with 14 and had one of the highest win/loss ratio with a record of 14 wins and 5 losses. The season was cut short due to the strike that occurred, which left Dennis tied with the most wins with Pete Vuckovich in the American League (who was still one game short of Dennis). That year the Cy Young award was granted to Rollie Fingers, which was the first time in MLB history that the Cy Young Award was given to a relief pitcher.

In 1991 and 1992 while playing with the Montreal Expos, Dennis had his lowest ERA of his career at 2.39 and 2.47 with both years having more than 220 innings pitched. In 1991, he had the lowest ERA of any started who had more than 200 innings and in 1992, he finished in the top 10 of lowest ERA with starters throwing more than 200 innings. In additiuon to highest ERA, he also contained an impressive record even though having an average of 3.5 for run support for games started, in comparison to Tom Glavine's 4.67 (whom went on to win the Cy Young with a record of 20-11)

```python
# number of wins with standard deviation of 5
max_num_wins = max(dennis_martinez_dataset.W)
std_dev_wins = dennis_martinez_dataset.W.std()
most_wins_years = dennis_martinez_dataset[
    dennis_martinez_dataset.W >= max_num_wins - std_dev_wins].sort_values(by=['W'], ascending=False)


best_era_years = dennis_martinez_dataset[
    dennis_martinez_dataset.ERA <= 
    min(dennis_martinez_dataset.ERA) + dennis_martinez_dataset.ERA.std()].sort_values(by=['ERA'], ascending=True)
```

```python
combined_stats = hof_pitching_starters.append(expos)
unique_player_ids = combined_stats.playerID.unique()
combined_stats['yearID'] = combined_stats[['yearID','yearID_x']].max(axis=1)
combined_stats = combined_stats.drop(['yearID_x'], axis=1)
```

```python
fig = go.Figure()

for player_id in unique_player_ids:
    stats = combined_stats[combined_stats.playerID == player_id]
    linewidth = 1.0
    if player_id == 'martide01':
        linewidth = 5.0
    fig.add_trace(go.Scatter(
        x=stats.yearID,
        y=stats.W,
        name=player_id,
        mode="lines+markers",
        line=dict(width=linewidth)
    ))

fig.update_layout(title="Number of Wins per Year", xaxis_title="# of Wins", yaxis_title="Year")
fig.show()
```

```python
fig = go.Figure()

for player_id in unique_player_ids:
    stats = combined_stats[combined_stats.playerID == player_id]
    linewidth = 1.0
    if player_id == 'martide01':
        linewidth = 5.0
    fig.add_trace(go.Scatter(
        x=stats.yearID,
        y=np.log(stats.W),
        name=player_id,
        mode="lines+markers",
        line=dict(width=linewidth)
    ))
fig.update_layout(title="Logged Number of Wins per Year", xaxis_title="# of Wins", yaxis_title="Year")
fig.show()
```

```python
fig = go.Figure()

for player_id in unique_player_ids:
    stats = combined_stats[combined_stats.playerID == player_id]
    linewidth = 1.0
    if player_id == 'martide01':
        linewidth = 5.0
    fig.add_trace(go.Scatter(
        x=stats.yearID,
        y=(stats.IPouts/3),
        name=player_id,
        mode="lines+markers",
        line=dict(width=linewidth)
    ))
    
fig.update_layout(title="Innings Pitched", xaxis_title="Year", yaxis_title="# of Innings")
fig.show()
```
